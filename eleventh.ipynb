```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
def createCorpus(doc_names) -> list:
    corpus = []
    for name in doc_names:
        with open(name, "r", encoding="utf-8") as f:
            text = f.read()
            corpus.append(text)

    return corpus

names = ["testtext.txt", "npunkt.txt", "input.txt"]
corpus = createCorpus(names)
corpus
```

```python
# Создание TF-IDF векторизатора
vectorizer = TfidfVectorizer()

# Вычисление TF-IDF
tfidf_matrix = vectorizer.fit_transform(corpus)
```

```python
# Получение списка ключевых слов
feature_names = vectorizer.get_feature_names_out()

# Преобразование матрицы в удобный для чтения формат
dense = tfidf_matrix.todense()
tfidf_df = pd.DataFrame(dense, columns=feature_names)

tfidf_df
```
